```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

plt.rcParams["figure.dpi"] = 140
```

```python
def sent_split(text: str):
    text = re.sub(r"\s+", " ", text.strip())
    parts = re.split(r'(?<=[.!?])\s+(?=[A-Z"\'(\[])', text)
    return [p.strip() for p in parts if p.strip()]
```

The speeches are splitted into sentences using rule-based regular expression that splits the sentences based on the white space,capilatization.This makes sure that sentiment is computed at the sentence level and also avoiding common abbreviation errors.

```python
def sentiment_score(sentence: str) -> float:

    return float(TextBlob(sentence).sentiment.polarity)
```

```python
def compute_time_series(title: str, text: str) -> pd.DataFrame:
    sentences = sent_split(text)
    scores = np.array([sentiment_score(s) for s in sentences], dtype=float)

    idx = np.arange(1, len(scores) + 1)
    t_norm = (idx - 1) / max(1, len(scores) - 1)

    return pd.DataFrame({
        "speech": title,
        "sentence_idx": idx,
        "t_norm": t_norm,
        "sentiment": scores,
        "sentence": sentences
    })

dfs = [compute_time_series(title, text) for title, text in SPEECH_TEXTS.items()]
df = pd.concat(dfs, ignore_index=True)

df.head()
```

Since the speech begins with greetings it has sentiment as 0 which means neutral.As the speech proceeds when he talks abotu poverty,war etc we can see the negitives dips and positives dips.

```python
def summarize(df_speech: pd.DataFrame) -> dict:
    scores = df_speech["sentiment"].to_numpy()
    q = max(1, len(scores) // 4)

    start_mean = float(np.mean(scores[:q]))
    end_mean = float(np.mean(scores[-q:]))

    x = df_speech["t_norm"].to_numpy()
    slope = float(np.polyfit(x, scores, 1)[0]) if len(scores) > 1 else 0.0

    return {
        "n_sentences": int(len(scores)),
        "mean_start_quartile": start_mean,
        "mean_end_quartile": end_mean,
        "end_minus_start": end_mean - start_mean,
        "linear_slope_per_norm_time": slope,
        "overall_mean": float(np.mean(scores)),
        "overall_std": float(np.std(scores)),
        "min": float(np.min(scores)),
        "max": float(np.max(scores)),
    }

summary = (df.groupby("speech")
             .apply(lambda g: pd.Series(summarize(g)))
             .reset_index()
             .sort_values("speech"))

summary
```

```python
fig, axes = plt.subplots(nrows=len(summary), ncols=1, figsize=(8, 8), sharex=False)

if len(summary) == 1:
    axes = [axes]

for ax, speech in zip(axes, summary["speech"]):
    d = df[df["speech"] == speech]

    ax.plot(d["sentence_idx"], d["sentiment"])
    ax.axhline(d["sentiment"].mean(), linestyle="--", linewidth=1)

    ax.set_title(speech)
    ax.set_ylabel("Sentiment")

axes[-1].set_xlabel("Sentence index (time)")
plt.tight_layout()
plt.show()
```

```python
plt.figure()

for speech in summary["speech"]:
    d = df[df["speech"] == speech]
    plt.plot(d["t_norm"], d["sentiment"], label=speech)

plt.axhline(df["sentiment"].mean(), linestyle="--", linewidth=1)

plt.title("Sentiment time series (normalized time)")
plt.xlabel("Normalized time (0=start → 1=end)")
plt.ylabel("Sentiment score (TextBlob polarity)")
plt.legend()
plt.show()
```

```python
def turning_points(df_speech: pd.DataFrame, k: int = 3):
    top_pos = df_speech.nlargest(k, "sentiment")[["sentence_idx", "sentiment", "sentence"]]
    top_neg = df_speech.nsmallest(k, "sentiment")[["sentence_idx", "sentiment", "sentence"]]
    return top_pos, top_neg

for speech in summary["speech"]:
    d = df[df["speech"] == speech]
    top_pos, top_neg = turning_points(d, k=3)

    print("\n" + "="*90)
    print(speech)

    print("\nTop positive sentences:")
    display(top_pos.assign(sentence=top_pos["sentence"].str.slice(0, 200)))

    print("\nTop negative sentences:")
    display(top_neg.assign(sentence=top_neg["sentence"].str.slice(0, 200)))
```
